## GDP Per State 

```python
import matplotlib.pyplot as plt
state =['Maha','Raj','Kel','Kar','TN','Ap','MP']
GDP=[6000,900,580,300,5500,390,100]
plt.bar(state,GDP,color='red')
plt.title('GDP per State',color='blue')
plt.xlabel('State Name')
plt.ylabel('GDP in Million')
```

## Histogram


```python
import matplotlib.pyplot as plt
score=[45,64,78,98,75,90,45,87,64,38,99,33,59,22,77,87,67]
plt.hist(score,bins=5,color='purple',edgecolor='black')
plt.xlabel('Score Number')
plt.ylabel('Number of Student')
plt.title('Score of Distrubution Students')
plt.show()
```
