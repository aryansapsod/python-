### Data Visualization And Manipulation


```python
# plt.plot(x,y color='color name',linestyle='line_Style',maker='marker symbol',label='label name')
import matplotlib.pyplot as plt
months = [1,2,3,4]
sales =[1000,1500,1200,1800]
plt.plot(months,sales,color='red',linestyle='-',linewidth= 2,marker='o',label='2025 sales data')
# plt.show()
plt.xlabel('months')
plt.ylabel('sales per months')
plt.title('Monthly Data Sales Reports')
plt.legend(loc ='upper right')
plt.grid(color ='grey')
plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4],['m1','m2','m3','m4'])
plt.yticks()
plt.show()
```

```python
product=['A','B','C','D']
Sales=[100,200,1000,500]
plt.bar(product,Sales,color='orange',label='Sales 2025')
plt.xlabel('product')
plt.ylabel('Sales')
plt.legend(loc ='upper left')
plt.title('Product Sale Comparison')
```

```python
# plt.pie(values=['north':3000,'south':2000,'east';1000,''],labels=label,color=color_list,autopct=%1.1f%%)
regions =["North","South","East","West"]
revenue=[3000,2000,1500,1000]
plt.pie(revenue,labels=regions,autopct='%1.1f%%',colors=['gold','skyblue','lightgreen','coral'])
plt.title('Revenue Contribution By Region')
```

## Scatter Plot

```python
import matplotlib.pyplot as plt
hours_studies =[1,2,3,4,5,6,7,8]
exam_score =[50,55,60,65,70,75,80,85]

#scatter plot
plt.scatter(hours_studies,exam_score,color='green',marker='o',label='Student Data')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Relationship Between Study Time and Exam Score')
plt.legend()
plt.grid(True)

```

```python
import matplotlib.pyplot as plt

plt.scatter([1,2,3],[55,62,78],color='blue',label='class A')
plt.scatter([1,2,3],[50,60,70],color='orange',label='Class B')


plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Comparison of Two Classes')
plt.legend()
plt.grid(True)

```

## Subplot and Layout Adjustment

```python
#plt.sunplot(nrows, ncol, index)
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[10,25,15,25]

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('line Chart')
plt.subplot(1,2,2)

plt.bar(x,y)
plt.title('Bar Chart')
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [10,20,25,15]
#fig ax = plt.subplot(nrows , ncol, figsize=(width ,height))

fig, ax =plt.subplots(1,2, figsize=(10,5))

ax[0].plot(x,y,color='purple')
ax[0].set_title('Line Plot')

ax[1].bar(x,y,color='purple')
ax[1].set_title('Bar Chart')
plt.tight_layout()

plt.show()
```
